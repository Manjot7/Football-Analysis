```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
pl_df = pd.read_csv('/content/Dataset/EPL_20_21.csv', encoding='latin-1')
pl_df.head()
```

```python
#Checking Non-Null Columns

pl_df.info()
```

```python
#Getting Stats on the Dataset

pl_df.describe()
```

```python
#Checking Null Values

pl_df.isna().sum()
```

```python
#Creating 2 New Columns

pl_df['Avg_Mins_Per_Match'] = (pl_df['Mins'] / pl_df['Matches']).astype(float)
pl_df['Avg_Goals_Per_Match'] = (pl_df['Goals'] / pl_df['Matches']).astype(float)
pl_df.head()
```

```python
#Total Goals Scored in the Season

Total_Goals = pl_df['Goals'].sum()
print(Total_Goals)
```

```python
#Total Penalty Goals

Total_PenGoals = pl_df['Penalty_Goals'].sum()
print(Total_PenGoals)
```

```python
#Total Penalty Attempts

Total_PenAttempts = pl_df['Penalty_Attempted'].sum()
print(Total_PenAttempts)
```

```python
#Pie Chart for Penalty Goals vs Attempts

plt.figure(figsize=(14, 6))
pen_not_scored = Total_PenAttempts - Total_PenGoals
data = [pen_not_scored, Total_PenGoals]
labels = ['Penalties Missed', 'Penalties Scored']
color = sns.color_palette('Paired')
plt.pie(data, labels = labels, colors = color, autopct = '%.0f%%')
plt.show()
```

```python
#Unique Positions

pl_df['Position'].unique()
```

```python
#Total FW Players

pl_df[pl_df['Position'] == 'FW']
```

```python
#Different Nationalities

np.size((pl_df['Nationality'].unique()))
```

```python
#Nationalities with the Most Players

nationality = pl_df.groupby('Nationality').size().sort_values(ascending = False)
nationality.head(10).plot(kind = 'bar', figsize=(13,6), color = sns.color_palette('Set2'))
```

```python
#Clubs with the Most Players

pl_df['Club'].value_counts().nlargest(5).plot(kind = 'bar', color = sns.color_palette('rocket'))
```

```python
#Clubs with the Least Players

pl_df['Club'].value_counts().nsmallest(5).plot(kind = 'bar', color = sns.color_palette('mako'))
```

```python
#Players based on Age Group

Under20 = pl_df[pl_df['Age'] <= 20]
Age20_25 = pl_df[(pl_df['Age'] > 20) & (pl_df['Age'] <= 25)]
Age25_30 = pl_df[(pl_df['Age'] > 25) & (pl_df['Age'] <= 30)]
Above30 = pl_df[pl_df['Age'] > 30]
```

```python
x = np.array([Under20['Name'].count(), Age20_25['Name'].count(), Age25_30['Name'].count(), Above30['Name'].count()])
mylabels = ['<=20', '20-25', '25-30', '>30']
plt.title('Total Players in Age Groups', fontsize = 20)
plt.pie(x, labels = mylabels, autopct = '%.1f%%')
plt.show()
```

```python
#Under 20 Players in Each Club

Under20_EachClub = pl_df[pl_df['Age'] < 20]
Under20_EachClub['Club'].value_counts().plot(kind = 'bar', color = sns.color_palette('tab10'))
```

```python
#Under 20 Players in Manchester United

Under20_EachClub[Under20_EachClub['Club'] == 'Manchester United']
```

```python
#Average Age in Each Club

plt.figure(figsize=(12, 6))
sns.boxplot(x = 'Club', y='Age', data = pl_df)
plt.xticks(rotation = 90)
```

```python
num_player = pl_df.groupby('Club').size()
data= (pl_df.groupby('Club')['Age'].sum())/ num_player
data.sort_values(ascending = False)
```

```python
#Total Assists from Each Club

Assists_Per_Club = pd.DataFrame(pl_df.groupby('Club', as_index = False)['Assists'].sum())
sns.set_theme(style='whitegrid', color_codes = True)
ax = sns.barplot(x = 'Club', y = 'Assists', data = Assists_Per_Club.sort_values(by = 'Assists'), palette = 'hls')
ax.set_xlabel('Club', fontsize = 20)
ax.set_ylabel('Assists', fontsize = 20)
plt.xticks(rotation = 90)
plt.rcParams['figure.figsize'] = (20,9)
plt.title('Assists Per Club', fontsize = 20)
```

```python
#Top 10 Assists

top_assists = pl_df[['Name', 'Club', 'Assists', 'Matches']].nlargest(n=10, columns = 'Assists')
top_assists
```

```python
Goals_Per_Club = pd.DataFrame(pl_df.groupby('Club', as_index = False)['Goals'].sum())
sns.set_theme(style='whitegrid', color_codes = True)
ax = sns.barplot(x = 'Club', y = 'Goals', data = Goals_Per_Club.sort_values(by = 'Goals'), palette = 'crest')
ax.set_xlabel('Club', fontsize = 20)
ax.set_ylabel('Goals', fontsize = 20)
plt.xticks(rotation = 90)
plt.rcParams['figure.figsize'] = (20,9)
plt.title('Goals Per Club', fontsize = 20)
```

```python
#Top Goal Scorers

top_goal_scorers = pl_df[['Name', 'Club', 'Goals', 'Matches']].nlargest(n=10, columns = 'Goals')
top_goal_scorers
```

```python
#Goals Per Match

goals_per_match = pl_df[['Name', 'Avg_Goals_Per_Match', 'Matches', 'Goals']].nlargest(n=10, columns = 'Avg_Goals_Per_Match')
goals_per_match
```

```python
#Goals with/without Assists

plt.figure(figsize=(16,8))
assists = pl_df['Assists'].sum()
data = [Total_Goals - assists, assists]
labels = ['Goals without Assists', 'Assisted Goals']
color = sns.color_palette('Set1')
plt.pie(data, labels = labels, colors = color, autopct = '%.0f%%')
plt.show()
```

```python
#Players with Most Yellow Cards

pl_yellow = pl_df.sort_values(by = 'Yellow_Cards', ascending = False)[:10]
plt.figure(figsize=(18,6))
plt.title('Most Yellow Cards')
p = sns.barplot(x=pl_yellow['Name'], y = pl_yellow['Yellow_Cards'], label = 'Players', color = 'Yellow')
plt.ylabel('Number of Yellow Cards')
p.set_xticklabels(p.get_xticklabels())
```

```python
#Players with Most Red Cards

pl_red = pl_df.sort_values(by = 'Red_Cards', ascending = False)[:10]
plt.figure(figsize=(18,6))
plt.title('Most Red Cards')
p = sns.barplot(x=pl_red['Name'], y = pl_red['Red_Cards'], label = 'Players', color = 'Red')
plt.ylabel('Number of Red Cards')
p.set_xticklabels(p.get_xticklabels())
```
